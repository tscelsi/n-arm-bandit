```python
from engines import EngineSimple, EngineArgmax, EngineEpsilonGreedy, EngineUCB
T = 1000
simple = EngineSimple(T=T)
argmax = EngineArgmax(T=T)
e_greedy_0_1 = EngineEpsilonGreedy(T=T, epsilon=0.1)
e_greedy_0_01 = EngineEpsilonGreedy(T=T, epsilon=0.01)
ucb = EngineUCB(c=2)
```

```python
simple.run()
argmax.run()
e_greedy_0_1.run()
e_greedy_0_01.run()
ucb.run()
```

```python
import matplotlib.pyplot as plt
import numpy as np
_, ax = plt.subplots()
for e in [simple, argmax, e_greedy_0_1, e_greedy_0_01, ucb]:
    average_rewards = np.array(e.reward_history).mean(axis=1)
    ax.plot(average_rewards, label=str(e))
plt.legend()
plt.show()
```
